Q1. Instruction ( For Data Cleaning ) - Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column.

```python
car = car.dropna(axis=0, how='all') # to remove all null colum contains NaN
```

```python
car.head()
```

```python
car.tail()
```

```python
car.shape
```

```python
car.isnull().sum() #answer
```

```python
car['Cylinders'].fillna(car['Cylinders'].mean(), inplace = True)
```
